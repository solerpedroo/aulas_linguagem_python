Faça uma pesquisa sobre os métodos remove(), del() e pop(). Para que servem, qual a diferença entre eles?

Em Python, os métodos `remove()`, `pop()` e a instrução `del` são usados para remover elementos de listas, mas cada um funciona de maneira diferente. O método `remove()` elimina a **primeira ocorrência de um valor específico**, ou seja, você precisa informar qual valor deseja remover, e se ele não estiver presente, será gerado um erro (`ValueError`). Já o método `pop()` remove e **retorna o elemento de um índice específico** (ou o último, se nenhum índice for passado); ele é útil quando se quer utilizar o valor removido. A instrução `del` serve para **remover elementos por índice ou por fatia (slice)**, podendo inclusive apagar a lista inteira se usada sem especificar um índice. Enquanto `remove()` age com base no valor, `pop()` e `del` trabalham com posições.

Elabore um programa que leia um conjunto de vários valores inteiros e os coloque em 2 listas conforme forem pares ou ímpares (uma lista para números pares e outra lista para números ímpares). A leitura dos números é finalizada quando um número negativo é lido

```python
lista_par = [] 
lista_impar = []

while True:
    num_digitado = int(input("Digite um número: "))

    if num_digitado < 0:
        break

    if num_digitado % 2 == 0:
        lista_par.append(num_digitado)
    else: 
        lista_impar.append(num_digitado)
    
print(f"Lista par {lista_par}")
print(f"Lista ímpar {lista_impar}")
```

Elabore um programa que leia uma lista de no máximo 10 elementos reais, o programa deverá imprimir o maior e segundo maior elemento e suas respectivas posições na lista

```python
numeros = []

print("Digite até 10 números reais (ou digite 'parar' para encerrar):")

for i in range(10):
    entrada = input(f"Digite o {i+1}º número: ")
    
    if entrada.lower() == "parar":
        break
    
    try:
        num = float(entrada)
        numeros.append(num)
    except ValueError:
        print("Entrada inválida. Digite um número real ou 'parar'.")
        continue

if len(numeros) < 2:
    print("É necessário pelo menos 2 números para encontrar o maior e o segundo maior.")
else:
    maior = max(numeros)
    
    restantes = [n for n in numeros if n < maior]

    if not restantes:
        print("Todos os elementos são iguais. Não há segundo maior distinto.")
    else:
        segundo_maior = max(restantes)
        pos_maior = numeros.index(maior)
        pos_segundo = numeros.index(segundo_maior)

        print(f"Maior número: {maior} na posição {pos_maior}")
        print(f"Segundo maior número: {segundo_maior} na posição {pos_segundo}")
```

Foram anotadas as idades e alturas de 30 alunos. Faça um Programa que determine quantos alunos com mais de 13 anos possuem altura inferior à média de altura desses alunos.

```python
idades = []
alturas = []

print("Digite as idades e alturas dos 30 alunos:")

for i in range(30):
    print(f"\nAluno {i+1}:")
    idade = int(input("Idade: "))
    altura = float(input("Altura (em metros, ex: 1.65): "))
    
    idades.append(idade)
    alturas.append(altura)

# Calcula a média de altura
media_altura = sum(alturas) / len(alturas)

# Conta quantos têm mais de 13 anos e altura abaixo da média
contador = 0
for i in range(30):
    if idades[i] > 13 and alturas[i] < media_altura:
        contador += 1

print(f"\nNúmero de alunos com mais de 13 anos e altura inferior à média: {contador}")
```

Construa um programa que leia dois números inteiros: a e b e uma lista com N valores inteiros (N fornecido pelo usuário). O programa deverá imprimir quantos elementos da Lista pertencem ao intervalo [a;b]

```python
# Lê os limites do intervalo
a = int(input("Digite o valor de a (limite inferior): "))
b = int(input("Digite o valor de b (limite superior): "))

# Garante que a <= b
if a > b:
    a, b = b, a

# Lê a quantidade de elementos na lista
N = int(input("Quantos números você deseja inserir na lista? "))

# Lê os N valores e conta os que estão no intervalo [a; b]
contador = 0
for i in range(N):
    num = int(input(f"Digite o {i+1}º número: "))
    if a <= num <= b:
        contador += 1

print(f"\nQuantidade de elementos no intervalo [{a}; {b}]: {contador}")
```

Construa um programa que seja constituído de uma lista GAB de 10 elementos caracteres (esta lista pode ser constituída somente dos caracteres a, b, c, d e e). O programa irá ler o nome e a resposta de 10 alunos de uma turma e deverá imprimir a nota de cada aluno (considerando que cada questão vale 1,0 ponto). O programa deverá também imprimir a média da sala

```python
# Lê o gabarito
print("Digite o gabarito das 10 questões (letras de 'a' a 'e'):")
gabarito = []
for i in range(10):
    while True:
        resposta = input(f"Questão {i+1}: ").lower()
        if resposta in ['a', 'b', 'c', 'd', 'e']:
            gabarito.append(resposta)
            break
        else:
            print("Resposta inválida. Use apenas 'a', 'b', 'c', 'd' ou 'e'.")

notas = []

# Lê os dados dos 10 alunos
for i in range(10):
    print(f"\nAluno {i+1}:")
    nome = input("Nome: ")
    respostas_aluno = []

    for j in range(10):
        while True:
            resposta = input(f"Resposta da questão {j+1}: ").lower()
            if resposta in ['a', 'b', 'c', 'd', 'e']:
                respostas_aluno.append(resposta)
                break
            else:
                print("Resposta inválida. Use apenas 'a', 'b', 'c', 'd' ou 'e'.")

    # Corrigir prova
    nota = 0
    for r_aluno, r_gabarito in zip(respostas_aluno, gabarito):
        if r_aluno == r_gabarito:
            nota += 1

    notas.append(nota)
    print(f"Nota de {nome}: {nota:.1f}")

# Média da sala
media = sum(notas) / len(notas)
print(f"\nMédia da turma: {media:.2f}")
```

Utilizando listas faça um programa que faça 5 perguntas para uma pessoa sobre um crime. As perguntas são:

a. "Telefonou para a vítima?"

b. "Esteve no local do crime?"

c. "Mora perto da vítima?"

d. "Devia para a vítima?"

e. "Já trabalhou com a vítima?"

O programa deve no final emitir uma classificação sobre a participação da pessoa no crime. Se a pessoa responder positivamente a 2 questões ela deve ser classificada como "Suspeita", entre 3 e 4 como "Cúmplice" e 5 como "Assassino". Caso contrário, ele será classificado como "Inocente".

```python
# Lista com as perguntas sobre o crime
perguntas = [
    "Telefonou para a vítima?",
    "Esteve no local do crime?",
    "Mora perto da vítima?",
    "Devia para a vítima?",
    "Já trabalhou com a vítima?"
]

# Lista para armazenar as respostas
respostas = []

# Faz as 5 perguntas e armazena as respostas
for pergunta in perguntas:
    resposta = input(pergunta + " (s/n): ").strip().lower()
    while resposta not in ['s', 'n']:
        resposta = input("Resposta inválida. Por favor, responda 's' ou 'n'. " + pergunta + " (s/n): ").strip().lower()
    respostas.append(resposta)

# Conta quantas respostas foram 'sim'
contador_sim = respostas.count('s')

# Classificação com base no número de respostas positivas
if contador_sim == 2:
    classificacao = "Suspeita"
elif 3 <= contador_sim <= 4:
    classificacao = "Cúmplice"
elif contador_sim == 5:
    classificacao = "Assassino"
else:
    classificacao = "Inocente"

# Exibe a classificação
print(f"\nClassificação: {classificacao}")
```
