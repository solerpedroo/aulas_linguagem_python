Elabore um programa que leia 2 notas e calcule a média

```python
a = int(input("Digite a primeira nota: "))
b = int(input("Digite a segunda nota: "))

media = (a+b)/2

print(f"A média de {a} e {b} é igual a {media}")
```

Elabore um programa que leia um número inteiro e imprima o seu antecessor e o seu sucessor

```python
numero_int = int(input("Digite um número inteiro: "))

antecessor = numero_int - 1
sucessor = numero_int + 1

print(f"O sucessor de {numero_int} é {sucessor} e o antecessor de {numero_int} é {antecessor}")
```

Elabore um programa que dado o lado de um quadrado determine a sua área e seu perímetro

```python
lado_quadrado = float(input("Digite o lado do quadrado: "))

area = lado_quadrado ** 2
perimetro = lado_quadrado * 4

print(f"A área do quadrado é {area} e o perímetro é {perimetro}")
```

Elabore um programa que imprima a parte inteira e o resto da divisão de dois números inteiros

```python
dividendo = float(input("Digite o dividendo: "))
divisor = float(input("Digite o divisor: "))

p_inteira = dividendo / divisor
resto = dividendo % divisor

print(f"A parte inteira da divisão é {p_inteira} e o resto é {resto}")
```
