Elabore um algoritmo que leia os valores reais de A, B, C e em seguida imprima na tela a soma entre A e B e mostre se a soma é menor que C

```python
a = float(input("Digite o valor A: "))
b = float(input("Digite o valor B: "))
c = float(input("Digite o valor C: "))

soma = a + b 

if soma < c:
    print(f"A soma entre {a} e {b} é igual a {soma} e é menor que o valor C {c}")
elif soma > c: 
    print(f"A soma entre {a} e {b} é igual a {soma} e é maior que o valor C {c}")
else:
    print(f"A soma entre {a} e {b} é igual a {soma} e é igual o valor C {c}")
```

Faça um algoritmo que calcule e escreva a média aritmética dos números inteiros entre 15 (inclusive) e 100 (inclusive)

```python
soma = 0
quantidade = 0

inicio = 15
fim = 100

for i in range (inicio, fim +1):
    soma += i
    quantidade += 1

media = soma / quantidade

print(media)
```

Faça um algoritmo que leia três valores inteiros diferentes e imprima na tela os valores em ordem decrescente

```python
a = int(input("Digite o valor A: "))
b = int(input("Digite o valor B: "))
c = int(input("Digite o valor C: "))

if a > b and a > c:
    if b > c:
        maior, meio, menor = a, b, c
    else: 
        maior, meio, menor = a, c, b
elif b > a and b > c: 
    if a > c:
        maior, meio, menor = b, a, c
    else: 
        maior, meio, menor = b, c, a
elif c > a and c > b:
    if a > b:
        maior, meio, menor = c, a, b
    else: 
        maior, meio, menor = c, b, a

print(f"A ordem descrente é igual a {maior, meio, menor}")
```

A jornada de trabalho semanal de um funcionário é de 40 horas. O funcionário que trabalhar mais de 40 horas receberá hora extra, cujo cálculo é o valor da hora regular com um acréscimo de 50%. Escreva um algoritmo que leia o número de horas trabalhadas em um mês, o salário por hora e escreva o salário total do funcionário, que deverá ser acrescido das horas extras, caso tenham sido trabalhadas

```python
salario_hora = float(input("Digite o seu salário recebido por hora: "))
horas_trabalhadas = int(input("Digite a quantidade de horas trabalhadas no mês: "))

horas_regulares = 40

if horas_trabalhadas > 40:
    horas_extras = (horas_trabalhadas - horas_regulares)
    salario_horas_extras = ((horas_extras) * ((salario_hora * 0.5) + salario_hora))
    salario_total_com_hr_extra = (salario_hora * horas_regulares) + salario_horas_extras
    print(f"O seu salário total com o acréscimo de horas extras é igual a R$ {salario_total_com_hr_extra}")
else: 
    salario_total_sem_hr_extra = salario_hora * horas_trabalhadas
    print(f"O seu salário total regular é de R$ {salario_total_sem_hr_extra}")
```

A prefeitura de uma cidade deseja fazer uma pesquisa entre seus habitantes. Faça um algoritmo para
coletar dados sobre o salário e número de filhos de cada habitante e após as leituras, escrever:

a) Média de salário da população

b) Média do número de filhos

c) Maior salário dos habitantes

d) Percentual de pessoas com salário menor que R$ 1500,00

Obs.: O final da leitura dos dados se dará com a entrada de um “salário negativo”

```python
qtd_habitantes = int(input("Insira a quantidade de habitantes da cidade: "))

salario_medio_habitantes = 0
qtd_media_filhos = 0

for i in range(qtd_habitantes):
    salario_habitante = float(input(f"Digite o salário do {i+1}º habitante: "))
    qtd_filhos = int(input(f"Insira a quantidade de filhos do {i+1}º habitante: "))

    if qtd_habitantes > 0:
        salario_medio_habitantes += salario_habitante
        qtd_media_filhos += qtd_filhos

if qtd_habitantes <= 0: 
    print(f"Não existe população nessa cidade")
else: 
    media_salario = salario_medio_habitantes / qtd_habitantes
    media_filhos = qtd_media_filhos / qtd_habitantes

    print(f"\n\tA média salarial nessa cidade é de R$ {media_salario} por habitante")
    print(f"\n\tA média de filhos por habitante na cidade é de {media_filhos}")
```

Dada a sequência de N elementos S= 1,2,5,12,29......(2*(N-1)+(N-2). Elabore:

a. Um programa que imprima o N-ésimo termo da sequência. O termo da sequência é fornecido pelo
usuário

b. Um programa que some os N elementos da sequência onde, N é fornecido pelo usuário

```python
#programa A
n = int(input("Insira o termo da sequência que você quer: "))

sequencia = (2*(n-1)+(n-2))

print(f"O {n}º termo da sequência é igual a {sequencia}")
```

```python
#programa B

n = int(input("Insira de quantidade de termos da sequência que você quer somar: "))

soma_sequencia = 0

for i in range(n): 
    sequencia = (2*(n-1)+(n-2))
    soma_sequencia += sequencia

print(soma_sequencia)
```

a. A média das Alturas das pessoas com idade entre 10 e 20 anos

b. A porcentagem das pessoas com peso inferior a 40 quilos entre todas as pessoas analisadas.

```python
qtd_pessoas_mais50 = 0
soma_altura_10a20anos = 0
qtd_altura_10a20anos = 0
qtd_peso_menor40kg = 0
peso_total = 0

for i in range(3):
    idade = int(input(f"Digite a idade da {i+1}ª pessoa: "))
    altura = int(input(f"Digite a altura da {i+1}ª pessoa: "))
    peso = int(input(f"Digite o peso da {i+1}ª pessoa: "))

    peso_total += peso

    if idade > 50:
        qtd_pessoas_mais50 += 1
    
    if  10 >= idade <= 20: 
        soma_altura_10a20anos += altura
        qtd_altura_10a20anos += 1

    if peso < 40: 
        qtd_peso_menor40kg += 1

if qtd_altura_10a20anos > 0:
    media_altura = soma_altura_10a20anos / qtd_altura_10a20anos
else:
    media_altura = 0

porcentagem_pessoas_menor_40kg = (qtd_peso_menor40kg / 3) * 100

print("\nResultados:")
print(f"Quantidade de pessoas com mais de 50 anos: {qtd_pessoas_mais50}")
print(f"Média das alturas das pessoas entre 10 e 20 anos: {media_altura:.2f} metros")
print(f"Porcentagem de pessoas com menos de 40 kg: {porcentagem_pessoas_menor_40kg:.2f}%")
 
```
