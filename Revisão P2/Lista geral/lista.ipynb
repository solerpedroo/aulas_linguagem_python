7. Elabore um programa que leia um número inteiro e construa duas funções: uma que some os
dígitos desse número inteiro e outra que determine o maior digito desse número. Exemplo:
número = 1063, então a soma =1+ 0 +6 + 3 =10 e o maior dígito é 6

```python
def soma_digitos(n):
    soma = 0
    n_str = str(n)
    for digito in n_str:
        if digito.isdigit():
            soma += int(digito)
    return soma

def maior_digito(n):
    maior = 0
    n_str = str(n)
    for digito in n_str:
        if digito.isdigit():
            if int(digito) > maior:
                maior = int(digito)
    return maior

numero = int(input("Digite um número inteiro: "))

print(f"Soma dos dígitos: {soma_digitos(numero)}")
print(f"Maior dígito: {maior_digito(numero)}")
```

8. Um número é perfeito se a soma dos seus divisores, com exceção dele mesmo, é igual a ele.
Exemplo: número 6, os divisores de 6 são 1, 2, 3 e 6. Somando-se 1 + 2 + 3 =6. Portanto 6 é um
número perfeito. Elabore um programa que leia um número e usando uma função determine se
ele é perfeito

```python
def num_perfeito():
    soma_divisores = 0 
    n = int(input("Digite um número: "))

    for i in range (1, n):
        if n % i == 0:
            soma_divisores += i

    print(f"A soma dos divisores é igual a {soma_divisores}")

num_perfeito()
```

```python
def contaVogais(texto): 
    cont = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}
    for i in texto.lower():
        if i in cont:
            cont[i] += 1
    return cont

texto = input("Digite um texto para ser analisado: ")
contaVogais(texto)
```

22. Construa um programa que seja constituído de uma lista GAB de 10 elementos caracteres ( esta
lista pode ser constituída somente dos caracteres a, b, c, d e e). O programa irá ler o nome e a
resposta de 10 alunos de uma turma e deverá imprimir a nota de cada aluno (considerando que
cada questão vale 1,0 ponto). O programa deverá também imprimir a média da sala

```python
print("Digite o gabarito das 10 questões (letras de 'a' a 'e'):")
gabarito = []
for i in range(10):
    while True:
        resposta = input(f"Questão {i+1}: ").lower()
        if resposta in ['a', 'b', 'c', 'd', 'e']:
            gabarito.append(resposta)
            break
        else:
            print("Resposta inválida. Use apenas 'a', 'b', 'c', 'd' ou 'e'.")

notas = []

for i in range(10):
    print(f"\nAluno {i+1}:")
    nome = input("Nome: ")
    respostas_aluno = []

    for j in range(10):
        while True:
            resposta = input(f"Resposta da questão {j+1}: ").lower()
            if resposta in ['a', 'b', 'c', 'd', 'e']:
                respostas_aluno.append(resposta)
                break
            else:
                print("Resposta inválida. Use apenas 'a', 'b', 'c', 'd' ou 'e'.")

    nota = 0
    for r_aluno, r_gabarito in zip(respostas_aluno, gabarito):
        if r_aluno == r_gabarito:
            nota += 1

    notas.append(nota)
    print(f"Nota de {nome}: {nota:.1f}")

media = sum(notas) / len(notas)
print(f"\nMédia da turma: {media:.2f}")
```

```python
def lucro(estoque):
    lucro_total = 0

    for id_prod, info in estoque.items():
        preco_compra = info['preco_compra']
        preco_venda = info['preco_venda']
        quantidade = info['quantidade']
        lucro_individual = (preco_venda - preco_compra) * quantidade
        lucro_total += lucro_individual
        print(f"\nID Produto: {id_prod} - Lucro Individual: R${lucro_individual:.2f}")
    return lucro_total

estoque = {
    1: {'preco_compra': 10.0, 'preco_venda': 15.0, 'quantidade': 100},
    2: {'preco_compra': 20.0, 'preco_venda': 30.0, 'quantidade': 50},
    3: {'preco_compra': 5.0, 'preco_venda': 10.0, 'quantidade': 200}
}

lucro_total = lucro(estoque)

print(f"\n\033[1;32mO lucro total com o estoque é de R$ {lucro_total:.2f}\033[0m")
```
