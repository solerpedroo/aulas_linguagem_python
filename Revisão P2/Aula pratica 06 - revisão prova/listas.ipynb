Elabore um programa que leia uma lista de no máximo 20 elementos inteiros. Em seguida o programa deverá imprimir a quantidade de valores múltiplos de 3

```python
lista_int = []
lista_mult3 = []

for i in range(20):
    num = int(input(f"Digite o {i+1}º elemento: "))
    lista_int.append(num)

    if num % 3 == 0:
        lista_mult3.append(num)
        
print(f"Os números múltiplos de 3 são: {lista_mult3}")
```

Elabore um programa que leia uma lista de no máximo 10 elementos reais, o programa deverá imprimir o maior e segundo maior elemento e suas respectivas posições na lista

```python
lista = []

for i in range (10): 
    num = int(input(f"Digite o {i+1}º número: "))

    lista.append(num)

    if len(lista) < 2:
        print("É necessário pelo menos dois números")
    else:
        maior = lista[0]
        for n in lista[1:]:
            if n > maior:
                maior = n
        
        segundo_maior_encontrado = False
        segundo_maior = None
        for n in lista:
            if n != maior:
                if not segundo_maior_encontrado or n > segundo_maior:
                    segundo_maior = n
                    segundo_maior_encontrado = True
        if not segundo_maior_encontrado: 
            print("Todos os números são iguais")
        else:
            pos1 = lista.index(maior)
            pos2 = lista.index(segundo_maior)
            print(f"O maior número é {pos1} e o segundo maior é {pos2}")
```

Foram anotadas as idades e alturas de 30 alunos. Faça um Programa que determine quantos alunos com mais de 13 anos possuem altura inferior à média de altura desses alunos

```python
idades = []
alturas = []

print ("Digite a idade e a altura de 30 alunos: ")

for i in range(30):
    idade = int(input(f"Idade do {i+1}º aluno: "))
    altura = int(input(f"Altura do {i+1}º aluno: "))  
    idades.append(idade)
    alturas.append(altura)

soma_alturas = 0

for altura in alturas:
    soma_alturas += altura
media_altura = soma_alturas / len(alturas)

cont = 0

for i in range (30):
    if idades[i] > 13 and alturas[i] < media_altura:
        cont += 1

print(f"Quantidade de alunos que tem mais de 13 anos e altura menor que a média : {cont}")
```

Construa um programa que leia dois números inteiros: a e b e uma lista com N valores inteiros (N fornecido pelo usuário). O programa deverá imprimir quantos elementos da Lista pertencem ao intervalo [a;b]


```python
lista = []
a = int(input("Digite o valor de A (limite inferior): "))
b = int(input("Digite o valor de B (limite superior): "))

if a > b:
    a, b = b, a

n = int(input("Quantos elementos você quer na lista? "))

cont = 0

for i in range (n):
    num = int(input(f"Digite o {i + 1}º valor: "))
    if num > a and num < b:
        cont += 1
        lista.append(num)

print(f"A quantidade de números entre {a} e {b} é {cont}.")
print(f"Os números digitados presente no intervalo são: {lista}")
```

Construa um programa que seja constituído de uma lista GAB de 10 elementos caracteres (esta lista pode ser constituída somente dos caracteres a, b, c, d e e). O programa irá ler o nome e a resposta de 10 alunos de uma turma e deverá imprimir a nota de cada aluno (considerando que cada questão vale 1,0 ponto). O programa deverá também imprimir a média da sala


```python
print("Digite o gabarito das 10 questões (letras de 'a' a 'e'):")
gabarito = []
for i in range(10):
    while True:
        resposta = input(f"Questão {i+1}: ").lower()
        if resposta in ['a', 'b', 'c', 'd', 'e']:
            gabarito.append(resposta)
            break
        else:
            print("Resposta inválida. Use apenas 'a', 'b', 'c', 'd' ou 'e'.")

notas = []

for i in range(10):
    print(f"\nAluno {i+1}:")
    nome = input("Nome: ")
    respostas_aluno = []

    for j in range(10):
        while True:
            resposta = input(f"Resposta da questão {j+1}: ").lower()
            if resposta in ['a', 'b', 'c', 'd', 'e']:
                respostas_aluno.append(resposta)
                break
            else:
                print("Resposta inválida. Use apenas 'a', 'b', 'c', 'd' ou 'e'.")

    nota = 0
    for r_aluno, r_gabarito in zip(respostas_aluno, gabarito):
        if r_aluno == r_gabarito:
            nota += 1

    notas.append(nota)
    print(f"Nota de {nome}: {nota:.1f}")

media = sum(notas) / len(notas)
print(f"\nMédia da turma: {media:.2f}")
```
